```python
#Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user
number = int(input("Enter a number to see if even or odd: "))
x=number%2
if (x) == 0:
   print("{0} is even".format(number))
else:
   print("{0} is odd".format(number))
```

```python
#EXTRAS
#1. If the number is a multiple of 4, print out a different message.
number = int(input("Enter a number: "))

if number%4==0:
  print(number, "is a multiple of 4")
elif number%2==0:
  print("{0} is even".format(number))
else:
   print("{0} is odd".format(number))
```

```python
#b. Ask the user for two numbers: one number to check (call it num) and one number to divide by (check). If check divides evenly into num, tell that to the user. If not, print a different appropriate message.
num=int(input("Enter a number: "))
check=int(input("Enter another number: "))
if num % check ==0:
  print(num, "divides evenly by", check)
else:
  print(num, "doesn't divide evenly by", check)
```
